```python
import pandas as pd
import numpy as np
```

```python
london_listings = pd.read_csv("../data/raw/london/listings.csv.gz")
london_calendar = pd.read_csv("../data/raw/london/calendar.csv.gz")
london_reviews = pd.read_csv("../data/raw/london/reviews.csv.gz")
london_neighbourhoods = pd.read_csv("../data/raw/london/neighbourhoods.csv")

ny_listings = pd.read_csv("../data/raw/new_york/listings.csv.gz")
ny_calendar = pd.read_csv("../data/raw/new_york/calendar.csv.gz")
ny_reviews = pd.read_csv("../data/raw/new_york/reviews.csv.gz")
ny_neighbourhoods = pd.read_csv("../data/raw/new_york/neighbourhoods.csv")
```

```python
datasets = {
    "London Listings": london_listings,
    "London Calendar": london_calendar,
    "London Reviews": london_reviews,
    "London Neighbourhoods": london_neighbourhoods,
    "NY Listings": ny_listings,
    "NY Calendar": ny_calendar,
    "NY Reviews": ny_reviews,
    "NY Neighbourhoods": ny_neighbourhoods
}

for name, df in datasets.items():
    print(f"{name}: {df.shape}")
```

```python
def schema_report(df, dataset_name):
    report = pd.DataFrame({
        "Column": df.columns,
        "Data Type": df.dtypes.astype(str).values,
        "Missing Values": df.isnull().sum().values,
        "Missing %": round((df.isnull().sum() / len(df)) * 100, 2).values,
        "Unique Values": [df[col].nunique() for col in df.columns]
    })

    print(f"\n{'='*60}")
    print(dataset_name)
    print('='*60)

    display(report)

    return report
```

```python
london_listing_schema = schema_report(
    london_listings,
    "London Listings"
)

ny_listing_schema = schema_report(
    ny_listings,
    "New York Listings"
)
```
