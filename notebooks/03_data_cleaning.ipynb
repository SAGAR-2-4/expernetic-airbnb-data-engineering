
# # Notebook 03 - Data Cleaning & Transformation
# Objective: Create clean listing datasets for London and New York.

```python
import pandas as pd
import numpy as np
```

```python
london_listings = pd.read_csv("../data/raw/london/listings.csv.gz")
ny_listings = pd.read_csv("../data/raw/new_york/listings.csv.gz")

london_clean = london_listings.copy()
ny_clean = ny_listings.copy()
```

```python
def clean_price_column(df, source_col="price", target_col="price_clean"):
    df[target_col] = (
        df[source_col]
        .str.replace("$", "", regex=False)
        .str.replace(",", "", regex=False)
        .astype(float)
    )
    return df
```

```python
london_clean = clean_price_column(london_clean)
ny_clean = clean_price_column(ny_clean)

print(london_clean[["price", "price_clean"]].head())
print(ny_clean[["price", "price_clean"]].head())
```

```python
london_clean["price_outlier"] = london_clean["price_clean"] > 1100
ny_clean["price_outlier"] = ny_clean["price_clean"] > 1562

print("London price outliers:", london_clean["price_outlier"].sum())
print("NY price outliers:", ny_clean["price_outlier"].sum())
```

```python
date_cols = [
    col for col in london_clean.columns
    if "date" in col.lower()
]

print(date_cols)
```

```python
date_cols_ny = [
    col for col in ny_clean.columns
    if "date" in col.lower()
]

print(date_cols_ny)
```

```python
possible_date_cols_london = [
    col for col in london_clean.columns
    if (
        col.endswith("_date")
        or col.endswith("_scraped")
        or col in ["host_since", "first_review", "last_review"]
    )
]

possible_date_cols_ny = [
    col for col in ny_clean.columns
    if (
        col.endswith("_date")
        or col.endswith("_scraped")
        or col in ["host_since", "first_review", "last_review"]
    )
]

print("London:", possible_date_cols_london)
print("NY:", possible_date_cols_ny)
```

```python
for col in possible_date_cols_london:
    london_clean[col] = pd.to_datetime(london_clean[col], errors="coerce")

for col in possible_date_cols_ny:
    ny_clean[col] = pd.to_datetime(ny_clean[col], errors="coerce")
```

```python
print("London date dtypes:")
print(london_clean[possible_date_cols_london].dtypes)

print("\nNY date dtypes:")
print(ny_clean[possible_date_cols_ny].dtypes)
```

```python
london_clean["host_tenure_years"] = (
    (london_clean["last_scraped"] - london_clean["host_since"]).dt.days / 365
)

ny_clean["host_tenure_years"] = (
    (ny_clean["last_scraped"] - ny_clean["host_since"]).dt.days / 365
)
```

```python
print("London host tenure:")
print(london_clean["host_tenure_years"].describe())

print("\nNY host tenure:")
print(ny_clean["host_tenure_years"].describe())
```
